```python
import sqlite3
import pandas as pd

conn = sqlite3.connect("sanity_check.db")
df = pd.read_sql("SELECT * FROM customer_summary", conn)

milk_pct = df["boughtMilk"].mean() * 100
cereal_given_milk = df[df.boughtMilk == 1]["boughtCereal"].mean() * 100
cereal_no_milk = df[df.boughtMilk == 0]["boughtCereal"].mean() * 100

print(milk_pct, cereal_given_milk, cereal_no_milk)
```
